```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/cricket_data_2025.csv')
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.sort_values(by='Year',inplace=True)
```

```python
df['Player_Name'].unique()
```

```python
df['Matches_Batted'].unique()
```

```python
df['Matches_Batted']=df['Matches_Batted'].astype(float)
```

```python
df['Not_Outs'].unique()
```

```python
df['Runs_Scored'].unique()
```

```python
df['Highest_Score'].unique()
```

```python
df['Highest_Score']=df['Highest_Score'].str.replace('*','').astype(float)
```

```python
for column in df.columns:
    unique_values = df[column].unique()  # Get unique values for the column

    # Convert all unique values to strings and join them with commas
    unique_values_str = ', '.join(map(str, unique_values))

    # Print the column name and its unique values in a single line
    print(f"Unique values in column '{column}': {unique_values_str}")
    print("-" * 50)  # Separator for readability
```

```python
df.columns
cols = ['Matches_Batted', 'Not_Outs', 'Runs_Scored',
       'Highest_Score', 'Batting_Average', 'Balls_Faced',
       'Batting_Strike_Rate', 'Centuries', 'Half_Centuries', 'Fours', 'Sixes',
       'Catches_Taken', 'Stumpings', 'Matches_Bowled', 'Balls_Bowled',
       'Runs_Conceded', 'Wickets_Taken',
       'Bowling_Average', 'Economy_Rate', 'Bowling_Strike_Rate',
       'Four_Wicket_Hauls', 'Five_Wicket_Hauls']
for col in cols:
  df[cols]=df[cols].astype(float)
```

```python
df.info()
```

```python
df.drop(labels='Best_Bowling_Match',axis=1,inplace=True)
```

```python
df.info()
```

```python
# Sort the DataFrame by 'Highest_Score' in descending order and select the top 5
top_5 = df.sort_values(by='Highest_Score', ascending=False).head(5)
top_5
```

```python
sns.barplot(data=top_5,x='Player_Name',y='Highest_Score')
```

```python
#top 5 most catches
top5_catches=df.sort_values(by='Catches_Taken', ascending=False).head(5)
sns.barplot(data=top5_catches,x='Player_Name',y='Catches_Taken')
```

```python
#top 5 Wickets Taken
top5_catches=df.sort_values(by='Wickets_Taken', ascending=False).head(5)
sns.barplot(data=top5_catches,x='Player_Name',y='Wickets_Taken')
```
