```python
import pandas as pd
import re

# Sample DataFrame
data = df_copy

df_copy = pd.DataFrame(data)

# Function to extract numerical MP from front camera
def extract_front_camera(mp_str):
    match = re.search(r'(\d+\.?\d*)MP', mp_str)
    return float(match.group(1)) if match else 0  # Convert to float for easy comparison

# Function to extract multiple back camera details
def extract_back_camera(camera_str):
    camera_info = {
        'Main': 0,
        'Ultra-wide': 0,
        'Macro': 0,
        'Telephoto': 0,
        'Other': 0
    }
    
    # Extract all megapixel values and their types
    matches = re.findall(r'(\d+\.?\d*)MP\s*\((.*?)\)|(\d+\.?\d*)MP', camera_str)
    
    for match in matches:
        mp_value = float(match[0] or match[2])  # Extract MP value
        camera_type = match[1].lower() if match[1] else 'Other'  # Extract camera type
        
        if 'main' in camera_type:
            camera_info['Main'] += mp_value
        elif 'ultra' in camera_type or 'wide' in camera_type:
            camera_info['Ultra-wide'] += mp_value
        elif 'macro' in camera_type:
            camera_info['Macro'] += mp_value
        elif 'telephoto' in camera_type:
            camera_info['Telephoto'] += mp_value
        else:
            camera_info['Other'] += mp_value
    
    return camera_info['Main'], camera_info['Ultra-wide'], camera_info['Macro'], camera_info['Telephoto'], camera_info['Other']

# Apply front camera extraction
df_copy['Front MP'] = df_copy['Front Camera'].apply(extract_front_camera)

# Apply back camera extraction and create new columns
df_copy[['Main MP', 'Ultra-wide MP', 'Macro MP', 'Telephoto MP', 'Other MP']] = df_copy['Back Camera'].apply(lambda x: pd.Series(extract_back_camera(x)))

# Drop original columns
df_copy.drop(columns=['Front Camera', 'Back Camera'], inplace=True)

# Print updated DataFrame
df_copy
```
