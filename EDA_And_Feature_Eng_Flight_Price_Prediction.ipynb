```python
## Feature Engi
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
df
```

```python
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].isnull().sum()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4,np.nan:1})
```

```python
df.head(1)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_Hour'] = df['Duration'].str.extract(r'(\d+)h', expand=False).fillna(0).astype(int)
df['Duration_Min'] = df['Duration'].str.extract(r'(\d+)m', expand=False).fillna(0).astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Additional_Info']]).toarray()
```

```python
df2 = pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Additional_Info','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
# Drop the original categorical columns
df = df.drop(['Airline', 'Source', 'Additional_Info','Destination'], axis=1)
```

```python
# Concatenate the remaining data with the encoded data
df_final = pd.concat([df, df2], axis=1)
```
